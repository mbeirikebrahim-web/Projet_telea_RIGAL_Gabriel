```python
import os
import shapefile
import rasterio

# Chemins
base = "/content/drive/MyDrive/PFE_POLE_MB/yolo_dataset"
img_path = os.path.join(base, "images", "agafay_rgb.tif")
shp_path = os.path.join(base, "arbres_boxes.shp")
label_path = os.path.join(base, "labels", "agafay_rgb.txt")

# Lire le raster
with rasterio.open(img_path) as src:
    width = src.width
    height = src.height

    # Lire le shapefile
    sf = shapefile.Reader(shp_path)
    shapes = sf.shapes()

    lignes = []

    for shp in shapes:
        # ignorer les géométries vides
        if shp.shapeType == 0 or len(shp.points) == 0:
            continue

        # bornes du polygone
        minx, miny, maxx, maxy = shp.bbox

        # convertir coordonnées géographiques -> pixels
        row_top, col_left = src.index(minx, maxy)
        row_bottom, col_right = src.index(maxx, miny)

        # remettre dans le bon ordre
        x1 = min(col_left, col_right)
        x2 = max(col_left, col_right)
        y1 = min(row_top, row_bottom)
        y2 = max(row_top, row_bottom)

        # éviter les boîtes invalides
        if x2 <= x1 or y2 <= y1:
            continue

        # format YOLO
        x_center = ((x1 + x2) / 2) / width
        y_center = ((y1 + y2) / 2) / height
        box_w = (x2 - x1) / width
        box_h = (y2 - y1) / height

        # bloquer entre 0 et 1
        x_center = max(0, min(1, x_center))
        y_center = max(0, min(1, y_center))
        box_w = max(0, min(1, box_w))
        box_h = max(0, min(1, box_h))

        # classe 0 = arbre
        lignes.append(f"0 {x_center:.6f} {y_center:.6f} {box_w:.6f} {box_h:.6f}")

# enregistrer le fichier txt
with open(label_path, "w", encoding="utf-8") as f:
    f.write("\n".join(lignes))

print("Fichier créé :", label_path)
print("Nombre de boîtes YOLO :", len(lignes))
```

```python
from pathlib import Path
import rasterio
from rasterio.windows import Window
import numpy as np
from PIL import Image

# ===== chemins d'entrée =====
base_in = Path("/content/drive/MyDrive/PFE_POLE_MB/yolo_dataset")
img_in = base_in / "images" / "agafay_rgb.tif"
lbl_in = base_in / "labels" / "agafay_rgb.txt"

# ===== nouveau dataset en tuiles =====
base_out = Path("/content/drive/MyDrive/PFE_POLE_MB/yolo_dataset_tiles")

for p in [
    base_out / "images" / "train",
    base_out / "images" / "val",
    base_out / "labels" / "train",
    base_out / "labels" / "val",
]:
    p.mkdir(parents=True, exist_ok=True)

# taille de tuile
TILE = 1024
VAL_EVERY = 5   # 1 tuile sur 5 va dans val

# ===== lire les labels YOLO de l'image entière =====
boxes = []
with open(lbl_in, "r", encoding="utf-8") as f:
    for line in f:
        parts = line.strip().split()
        if len(parts) != 5:
            continue
        cls, xc, yc, bw, bh = parts
        boxes.append((int(cls), float(xc), float(yc), float(bw), float(bh)))

train_count = 0
val_count = 0
saved_total = 0

with rasterio.open(img_in) as src:
    W = src.width
    H = src.height

    # convertir labels YOLO -> boîtes absolues en pixels
    abs_boxes = []
    for cls, xc, yc, bw, bh in boxes:
        xc *= W
        yc *= H
        bw *= W
        bh *= H
        x1 = xc - bw / 2
        y1 = yc - bh / 2
        x2 = xc + bw / 2
        y2 = yc + bh / 2
        abs_boxes.append((cls, x1, y1, x2, y2))

    tile_id = 0

    for y in range(0, H - TILE + 1, TILE):
        for x in range(0, W - TILE + 1, TILE):

            current_labels = []

            for cls, x1, y1, x2, y2 in abs_boxes:
                xc = (x1 + x2) / 2
                yc = (y1 + y2) / 2

                # garder la boîte si son centre est dans la tuile
                if x <= xc < x + TILE and y <= yc < y + TILE:
                    lx1 = max(x1 - x, 0)
                    ly1 = max(y1 - y, 0)
                    lx2 = min(x2 - x, TILE)
                    ly2 = min(y2 - y, TILE)

                    bw = lx2 - lx1
                    bh = ly2 - ly1

                    if bw <= 1 or bh <= 1:
                        continue

                    xc_l = ((lx1 + lx2) / 2) / TILE
                    yc_l = ((ly1 + ly2) / 2) / TILE
                    bw_l = bw / TILE
                    bh_l = bh / TILE

                    current_labels.append(
                        f"{cls} {xc_l:.6f} {yc_l:.6f} {bw_l:.6f} {bh_l:.6f}"
                    )

            # ignorer les tuiles vides
            if len(current_labels) == 0:
                continue

            # lire seulement les 3 bandes RGB
            tile = src.read([1, 2, 3], window=Window(x, y, TILE, TILE))
            tile = np.transpose(tile, (1, 2, 0))
            tile = np.clip(tile, 0, 255).astype(np.uint8)

            split = "val" if (saved_total % VAL_EVERY == 0) else "train"
            name = f"agafay_{tile_id:04d}"

            Image.fromarray(tile).save(
                base_out / "images" / split / f"{name}.jpg",
                quality=95
            )

            with open(base_out / "labels" / split / f"{name}.txt", "w", encoding="utf-8") as f:
                f.write("\n".join(current_labels))

            if split == "train":
                train_count += 1
            else:
                val_count += 1

            saved_total += 1
            tile_id += 1

# créer data.yaml
yaml_text = f"""path: {base_out}
train: images/train
val: images/val

names:
  0: arbre
"""

with open(base_out / "data.yaml", "w", encoding="utf-8") as f:
    f.write(yaml_text)

print("Terminé.")
print("Train tiles :", train_count)
print("Val tiles   :", val_count)
print("Data yaml   :", base_out / "data.yaml")
```
